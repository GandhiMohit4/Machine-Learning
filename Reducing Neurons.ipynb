```
(x_train, y_train), (x_test, y_test) = mnist.load_data('/tmp/mnist.npz')
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

x_train /= 255
x_test /= 255

x_train = x_train.reshape(-1, 28*28)
x_test = x_test.reshape(-1, 28*28)
x_train.shape, x_test.shape, y_train, y_test
```

# Exercise
Make a dense neural network with 95%+ accuracy on Mnist that has the smallest number of neurons possible by experimenting with Dropout, and Batch Norm

```
from sklearn.model_selection import train_test_split
from numpy import dot

def train_model(nodeCnt_1,nodeCnt_2):
  
  model = Sequential()
  model.add(Dense(nodeCnt_1, activation = 'relu',input_shape=(784,)))
  model.add(Dense(nodeCnt_2, activation = 'relu'))
  model.add(Dense(10,activation = 'softmax'))
  
  model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
  
  return model
  
def reduce_nodes(x_train, y_train, x_test, y_test,nodeCnt_1,nodeCnt_2, learn_rate_1, learn_rate_2):
  
  x_train, x_val, y_train, y_val = train_test_split(x_train, y_train)
  
  min_results = [1, nodeCnt_1,nodeCnt_2]
  target_accuracy = 0.95000
  retryCount = 5
  
  while ((nodeCnt_1 > 0 and nodeCnt_2 > 0) or retryCount == 0):
    output_model = train_model(nodeCnt_1,nodeCnt_2)
    output_model.summary()
    
    h = output_model.fit(x_train, y_train, epochs=10, batch_size = 200,
              validation_data=(x_val, y_val))
    
    test_loss, test_accuracy = output_model.evaluate(x_test, y_test, batch_size=64)
    
    print('test accuracy: ', test_accuracy)
    
    if test_accuracy >= target_accuracy and ((nodeCnt_1 + nodeCnt_2)  < (min_results[1] + min_results[2])):
      min_results[0] = test_accuracy
      min_results[1] = nodeCnt_1
      min_results[2] = nodeCnt_2
      retryCount = 5
    else:
      retryCount -= 1
    
   
    nodeCnt_1 -= learn_rate_1
    nodeCnt_2 -= learn_rate_2
 
    
    
  print('min_results: ', min_results) 
  return min_results

```

```
nodeCnt_1 = 45
nodeCnt_2 = 45
learn_rate_1 = 2
learn_rate_2 = 2
min_results = reduce_nodes(x_train, y_train, x_test, y_test,nodeCnt_1,nodeCnt_2, learn_rate_1, learn_rate_2)
```
